## Loading the dataset

```python
import pandas as pd
world_dataset = pd.read_csv('/content/Online_Learning_Data.csv')
world_dataset
```

We will first extract the years from this dataset

```python
world_dataset_copy=world_dataset.copy()
world_dataset_copy['Date']=pd.to_datetime(world_dataset['Month']).dt.strftime('%m/%Y')
world_dataset_copy['year']=pd.to_datetime(world_dataset_copy['Date']).dt.year
```

```python
world_dataset_grouped=world_dataset_copy.groupby(pd.Grouper(key='year',axis=0)).mean()
world_dataset_grouped
```

Now lets transpose this dataset

```python
world_dataset_grouped_transpose=world_dataset_grouped.T
world_dataset_grouped_transpose
```

MAking the dataset fit for a scatter_map_box

```python

col1 = []
col2 = []
col3 = []
# I made three columns and appended the required data in the required format
for i, c in enumerate(list(world_dataset_grouped_transpose.index)):
    for j in world_dataset_grouped_transpose.columns[1:]:
        col1.append(c)
        col2.append(j)
        col3.append(world_dataset_grouped_transpose.iloc[i][j])

final_world_dataset = pd.DataFrame({"country": col1, "year": col2, "value": col3})
final_world_dataset
```

```python
final_world_dataset.to_csv('Preprocessed_world_dataset.csv')
```
